```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

## Line Plot

```python
flights = sns.load_dataset("flights")
flights
```

```python
may_flights = flights[flights["month"]=="May"]

plt.plot(may_flights["year"], may_flights["passengers"])
plt.xlabel("Year")
plt.ylabel("Passengers")
plt.title("Line Chart")
plt.show()


sns.lineplot(data=may_flights,x="year", y="passengers")
plt.title("Seaborn Charts")
plt.show()
```

```python
year_1950_flights = flights[flights["year"]==1959]
plt.plot(year_1950_flights["month"],year_1950_flights["passengers"])
plt.xlabel("Month")
plt.ylabel("Number of passengers")
plt.title("Year 1950 flight data:")
plt.show()

```

```python
years = flights['year'].unique()
total_paassengers = []
for year in flights['year'].unique():
    flight_year = flights[flights['year']==year]
    total = flight_year['passengers'].sum()
    print(f"year:{year}, sum:{total}")
    total_paassengers.append(total)
```

```python
#Year wise total passengers line plot


plt.plot(years,total_paassengers)
plt.xlabel("Year")
plt.ylabel("Number of passengers")
plt.title("Year flight data:")
plt.show()
```

```python
#Using pandas groupby to group all the flights data for same year
years2 = flights.groupby(by='year').aggregate({'passengers':"sum"})
years2
```

```python
years3 = flights.groupby(by='year').aggregate({'passengers':"sum","month":"first"})
years3
```

```python
#Using groupby month 
months_groupby = flights.groupby(by="month").aggregate({"passengers":"sum"})
months_groupby
```

```python
#Poting line chart for months and passengers

sns.lineplot(data=months_groupby,x="month", y="passengers")
plt.title("Months Charts")
plt.show()
```

```python
years2 = flights.groupby(by='year').aggregate({'passengers':"sum"})
years2
```

```python
taxies = sns.load_dataset("taxis")
taxies
```

```python
credit_card_users = taxies[taxies["payment"]=="credit card"]    
credit_card_users = credit_card_users.sort_values(by=['distance'])

plt.plot(credit_card_users["distance"], credit_card_users["total"])
plt.xlabel("Distance")
plt.ylabel("Fares")
plt.title("Line Chart")
plt.show()
```

```python
credit_card_users.sort_values(by=['total'],ascending=False)
credit_card_users
```

```python
cash_users = taxies[taxies["payment"] == "cash"]
cash_users = cash_users.sort_values(by=['distance'])


plt.plot(cash_users["distance"], cash_users["total"])
plt.xlabel("Distance")
plt.ylabel("Fares")
plt.title("Line Chart")
plt.show()
```
