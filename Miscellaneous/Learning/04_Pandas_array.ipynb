## Cleaning 

After analysis of the data we have  is not realised that our dataset is not perfect, some columns names are irrelevant,there are a few missing data , some

```python
cars = pd.read_csv("mtcars.csv")
cars
```

```python
#Renaming the Column
cars=cars.rename(columns={ 'model':'Model Name' })
cars
```

```python
cars.describe()
```

```python
cars.info
```

```python
#Filling the null Values 
cars.qsec = cars.qsec.fillna(cars.qsec.mean())
cars
```
