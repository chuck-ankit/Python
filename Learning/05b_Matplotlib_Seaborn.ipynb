## Line Plot

```python
flights = sns.load_dataset("flights")
flights
```

```python
may_flights = flights[flights["month"]=="May"]

plt.plot(may_flights["year"], may_flights["passengers"])
plt.xlabel("Year")
plt.ylabel("Passengers")
plt.title("Line Chart")
plt.show()


sns.lineplot(data=may_flights,x="year", y="passengers")
plt.title("Seaborn Charts")
plt.show()
```

```python
year_1950_flights = flights[flights["year"]==1959]
plt.plot(year_1950_flights["month"],year_1950_flights["passengers"])
plt.xlabel("Month")
plt.ylabel("Number of passengers")
plt.title("Year 1950 flight data:")
plt.show()

```

```python
years = flights['year'].unique()
total_paassengers = []
for year in flights['year'].unique():
    flight_year = flights[flights['year']==year]
    total = flight_year['passengers'].sum()
    print(f"year:{year}, sum:{total}")
    total_paassengers.append(total)
```

```python
#Year wise total passengers line plot


plt.plot(years,total_paassengers)
plt.xlabel("Year")
plt.ylabel("Number of passengers")
plt.title("Year flight data:")
plt.show()
```

```python
#Using pandas groupby to group all the flights data for same year
years2 = flights.groupby(by='year').aggregate({'passengers':"sum"})
years2
```

```python
years3 = flights.groupby(by='year').aggregate({'passengers':"sum","month":"first"})
years3
```

```python
#Using groupby month 
months_groupby = flights.groupby(by="month").aggregate({"passengers":"sum"})
months_groupby
```

```python
#Poting line chart for months and passengers

sns.lineplot(data=months_groupby,x="month", y="passengers")
plt.title("Months Charts")
plt.show()
```

```python
years2 = flights.groupby(by='year').aggregate({'passengers':"sum"})
years2
```

## Find out ratio of the class who survived age 20 to 30 in titanic crash? 

```python
df
age_20_30 = df[(df.age >=20 ) & (df.age <= 30) & (df["survived"]==1)]

filtered_data = age_20_30.groupby(by='class').agg('count')
filtered_data = filtered_data.reset_index()
filtered_data
```

```python
#ploting the ratio of the class 
sns.barplot(data=filtered_data,x=filtered_data['class'],y=filtered_data["survived"])
```

```python
## Finding out which sex survived the most in the age of 20 to 30
sex_survivial_rate = age_20_30.groupby(by='sex').agg('count')
sex_survivial_rate = sex_survivial_rate.reset_index()
#Ploting bar chart
sns.barplot(data = sex_survivial_rate,x=sex_survivial_rate['sex'],y = sex_survivial_rate['survived'])
plt.show()
```

Parameters:

Input data structure. Either a long-form collection of vectors that can be assigned to named variables or a wide-form dataset that will be internally reshaped.

Grouping variable that will produce points with different colors. Can be either categorical or numeric, although color mapping will behave differently in latter case.

Grouping variable that will produce points with different markers. Can have a numeric dtype but will always be treated as categorical.

Method for choosing the colors to use when mapping the hue semantic. String values are passed to color_palette(). List or dict values imply categorical mapping, while a colormap object implies numeric mapping.

Specify the order of processing and plotting for categorical levels of the hue semantic.

hue_normtuple or matplotlib.colors.Normalize
Either a pair of values that set the normalization range in data units or an object that will map from data units into a [0, 1] interval. Usage implies numeric mapping.

sizeslist, dict, or tuple
An object that determines how sizes are chosen when size is used. List or dict arguments should provide a size for each unique data value, which forces a categorical interpretation. The argument may also be a min, max tuple.

size_orderlist
Specified order for appearance of the size variable levels, otherwise they are determined from the data. Not relevant when the size variable is numeric.

size_normtuple or Normalize object
Normalization in data units for scaling plot objects when the size variable is numeric.

markersboolean, list, or dictionary
Object determining how to draw the markers for different levels of the style variable. Setting to True will use default markers, or you can pass a list of markers or a dictionary mapping levels of the style variable to markers. Setting to False will draw marker-less lines. Markers are specified as in matplotlib.

Histogram -> to check the distribution of data.
          -> has to convert from continuous to categorical.
          -> deals with only one variable.
          
          
- sorts the list.
- it puts data into bins.(this is decided by the user.)
- then counts value for each bin.(bin is category)





```python
penguins = sns.load_dataset("penguins")
penguins
```

```python
## Creating Histogram for penguins at iland Biscoe and bill length using matplotlib

penguins_at_biscoe = penguins[penguins['island']=="Biscoe"]
b = plt.hist(penguins_at_biscoe['bill_length_mm'],bins=50)
print(b[1])
plt.xlabel("bill_lenght_mm")
plt.ylabel("Counts")
plt.title("Matplotlib Histogram ")
plt.show()
## count = frequency
## Creating Histogram for penguins at iland Biscoe and bill length using seaborn

sns.histplot(data=penguins_at_biscoe, x=penguins_at_biscoe['bill_length_mm'], bins=30, hue='sex')
plt.title("Seaborn Histogram")
plt.show()


sns.histplot(data=penguins_at_biscoe, x=penguins_at_biscoe['bill_length_mm'], bins=30, hue='species')
plt.title("Seaborn Histogram")
plt.show()
```
