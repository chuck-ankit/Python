## Crosstab(Pandas Function)
The crosstab() function is used to compute a simple cross tabulation of two (or more) factors.

```python
pd.crosstab(car_sales["EngineV"], car_sales["Brand"])
```

```python
brand_body = pd.crosstab(car_sales["Brand"], car_sales["Body"])
brand_body
```

## Groupby function in Pandas
The groupby function can be used to group data by any column or columns in a DataFrame. It is a powerful tool that can be used to perform a variety of tasks, such as aggregating data, calculating statistics, and filtering data.

```python
car_sales.groupby("Brand")
```

```python
# Create a DataFrame
df = pd.DataFrame({"name": ["Alice", "Bob", "Carol", "David"], "age": [25, 30, 35, 40]})

```

```python
# Group the DataFrame by the "age" column
grouped = df.groupby("age")
```

```python
mean_age = grouped.mean()
```

```python

# View the results
print(mean_age)
```
