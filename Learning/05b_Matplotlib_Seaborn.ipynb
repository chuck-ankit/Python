```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

## Line Plot

```python
flights = sns.load_dataset("flights")
flights
```

```python
may_flights = flights[flights["month"]=="May"]

plt.plot(may_flights["year"], may_flights["passengers"])
plt.xlabel("Year")
plt.ylabel("Passengers")
plt.title("Line Chart")
plt.show()


sns.lineplot(data=may_flights,x="year", y="passengers")
plt.title("Seaborn Charts")
plt.show()
```

```python
taxies = sns.load_dataset("taxis")
taxies
```

```python
credit_card_users = taxies[taxies["payment"]=="credit card"]    
credit_card_users = credit_card_users.sort_values(by=['distance'])

plt.plot(credit_card_users["distance"], credit_card_users["total"])
plt.xlabel("Distance")
plt.ylabel("Fares")
plt.title("Line Chart")
plt.show()
```

```python
credit_card_users.sort_values(by=['total'],ascending=False)
credit_card_users
```
