```python
import pandas as pd
from sklearn.linear_model import LinearRegression

df = pd.read_csv('https://raw.githubusercontent.com/krishnaik06/simple-Linear-Regression/master/Salary_Data.csv')
print(df)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt 
sns.scatterplot(data=df, x= "YearsExperience", y= "Salary" )
plt.show()
```

```python
years_Experieanc = df["YearsExperience"].values
salary = df["Salary"].values

years_Experieanc = years_Experieanc.reshape(-1,1)

model = LinearRegression()

model.fit(years_Experieanc, salary)

c = model.intercept_
m = model.coef_[0]

print(f"M:{m}, C:{c}")
```

```python
new_experience = 7

predicted_salary = model.predict([[new_experience]])

print(f"Predicted Salary for Experience:{new_experience} is {predicted_salary} ")
```
